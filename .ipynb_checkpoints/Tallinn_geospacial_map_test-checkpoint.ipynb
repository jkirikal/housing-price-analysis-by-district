```python
import geopandas as gpd

# Replace 'your_shapefile.shp' with the actual filename of your shapefile
gdf = gpd.read_file('tallinn_asumid/t02_41_asum.shp')
```

```python
gdf.columns
```

```python
import matplotlib.pyplot as plt

plt.title('Tallinn Districts')
gdf.plot(edgecolor='black', linewidth=0.5, column='linnaosa_n', legend=True, legend_kwds={'loc': 'upper left'}, figsize=(10, 10))
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

```
